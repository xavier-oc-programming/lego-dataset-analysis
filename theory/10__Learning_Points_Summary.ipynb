## **10)—Learning Points and Summary**

### **Overview**

This final section consolidates everything learned throughout **Day 74: Aggregate & Merge Data with Pandas – LEGO Dataset**. The lesson combined **data aggregation**, **visualisation**, and **relational data concepts** to analyse LEGO’s product catalogue over time.

---

### **Key Learning Points**

In this lesson, we learned how to:

* Use **HTML and Markdown** inside Jupyter Notebooks

  * Section headings using `#`, `##`, `###`
  * Embed images using the HTML `<img>` tag

* Aggregate data using Pandas:

  * Combine `.groupby()` with `.count()` to summarise data
  * Use `.value_counts()` to count categorical values
  * Apply `.agg()` to perform column-specific operations
  * Compute statistics such as:

    * counts
    * unique values
    * means

* Slice DataFrames using square bracket notation:

  * `df[:10]`
  * `df[:-2]`
  * Useful for excluding incomplete or forward-looking data

* Rename DataFrame columns using:

  * `.rename(columns=...)`

---

### **Data Visualisation Techniques Covered**

We created multiple types of charts using **Matplotlib**, including:

* **Line charts**

  * Used to visualise trends over time
  * Implemented dual y-axes (`twinx`) to compare values with different scales

* **Scatter plots**

  * Used to observe trends and dispersion
  * Applied to average LEGO set complexity over time

* **Bar charts**

  * Used to compare categorical data
  * Visualised the top LEGO themes by number of sets
  * Improved readability with:

    * figure sizing
    * axis labels
    * rotated tick labels

---

### **Relational Data Concepts**

We worked with LEGO data as a **relational dataset**, learning how to:

* Understand **primary keys**

  * Unique identifiers (e.g. `id` in `themes.csv`)

* Understand **foreign keys**

  * References to another table (e.g. `theme_id` in `sets.csv`)

* Merge tables using:

  * `pd.merge()` on a shared key column

This allowed us to:

* Combine theme names with set counts
* Answer real-world questions such as:

  * Which LEGO theme has the most sets?

By the end of this lesson, you should be comfortable:

* Moving from raw CSV files to meaningful insights
* Aggregating large datasets efficiently with Pandas
* Choosing the correct visualisation for your data
* Handling incomplete or misleading data responsibly
* Thinking about data in **relational terms**, not just flat tables

This day tied together **data analysis**, **visualisation**, and **data modelling**, making it one of the most comprehensive lessons so far.
