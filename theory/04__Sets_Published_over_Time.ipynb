### **Important Data Quality Observation**

The dataset was created in **late 2020**, meaning:

* 2020 is an incomplete year
* 2021 contains only a handful of forward-looking entries

If we plot all years as-is, the sharp drop at the end of the graph is **misleading**.

---

### **Initial Plot (Misleading)**

```python
plt.plot(sets_by_year.index, sets_by_year.set_num)
```

This produces an artificial decline that does **not** reflect LEGO’s actual performance.

---

### **Correcting the Plot Using Slicing**

To visualise only **complete calendar years**, we exclude the last two rows using Python slicing.

```python
plt.plot(
    sets_by_year.index[:-2],
    sets_by_year.set_num[:-2]
)
```

This technique applies standard Python list slicing to Pandas index and Series objects.

---

### **Interpreting the Trend**

From the corrected chart, we can observe:

* LEGO released **fewer than 10 sets per year** in its early decades
* Growth remained steady for roughly the first 45 years
* A **sharp acceleration** began in the mid-1990s
* A brief decline occurred in the early 2000s
* Strong recovery followed around 2005
* By 2019, LEGO released **over 800 sets in a single year**

---

### **Key Takeaways**

* `.groupby().count()` is ideal for yearly aggregation
* Always check for **incomplete time periods** in time-series data
* Python slicing (`[:-2]`) is a simple but powerful data-cleaning tool
* Visual trends can be misleading if raw data is plotted without context
