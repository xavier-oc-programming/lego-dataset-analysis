## 2)— **Solution: Exploring the LEGO Brick Colours**

### **Import Pandas**

Before working with the dataset, we import **Pandas**, which will be our main data analysis tool throughout the LEGO project.

```python
import pandas as pd
```

---

### **Reading the Colours Dataset**

We load the `colors.csv` file from the `data/` directory and inspect the first few rows to understand its structure.

```python
colors = pd.read_csv('data/colors.csv')
colors.head()
```

#### **Observed Columns**

* `id` → unique identifier for each colour
* `name` → colour name
* `rgb` → RGB hex value
* `is_trans` → transparency flag (`'t'` = transparent, `'f'` = opaque)

---

### **Counting Unique LEGO Colours**

To determine how many different colours LEGO produces, we count the number of unique entries in the `name` column.

```python
colors['name'].nunique()
```

#### **Result**

* LEGO produces **135 unique colours**

---

### **Finding Transparent vs Opaque Colours**

The dataset includes a column called `is_trans` that indicates whether a colour is transparent.

#### **Method 1: Using `groupby()` and `count()`**

```python
colors.groupby('is_trans').count()
```

This groups the dataset by transparency and counts the number of rows in each group.

---

#### **Method 2: Using `value_counts()` (simpler and faster)**

```python
colors.is_trans.value_counts()
```

#### **Result Interpretation**

* `'f'` → **107 opaque colours**
* `'t'` → **28 transparent colours**

---

### **Key Takeaways**

* `.nunique()` is ideal for counting unique categorical values
* `.groupby().count()` provides structured grouped summaries
* `.value_counts()` is the quickest way to count category frequencies
* LEGO colours are mostly opaque, with a smaller subset of transparent variants
