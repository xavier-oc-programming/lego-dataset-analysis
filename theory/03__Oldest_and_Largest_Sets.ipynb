### **Exploring the `sets.csv` Dataset**

The `sets.csv` file contains detailed information about individual LEGO sets.

We begin by loading the dataset and inspecting its structure.

```python
sets = pd.read_csv('data/sets.csv')
sets.head()
```

#### **Key Columns Identified**

* `set_num` → unique set identifier
* `name` → name of the LEGO set
* `year` → year the set was released
* `theme_id` → numeric identifier for the theme
* `num_parts` → number of parts in the set

To understand the full range of the dataset, we also inspect the last few rows.

```python
sets.tail()
```

---

### **Finding the First LEGO Sets Ever Released**

To identify the **earliest LEGO sets**, we sort the dataset by the `year` column in ascending order.

```python
sets.sort_values('year').head()
```

#### **Result**

* LEGO’s first sets were released in **1949**
* Early products included:

  * Extra-Large Gift Set (ABB)
  * Large Gift Set (ABB)
  * Medium Gift Set (ABB)
  * Small Brick Set (ABB)
  * Small Doors and Windows Set (ABB)

---

### **How Many Sets Did LEGO Sell in Its First Year?**

To find how many products LEGO sold in its launch year, we filter the dataset for `year == 1949`.

```python
sets[sets['year'] == 1949]
```

#### **Result**

* LEGO sold **5 different sets** in 1949

This demonstrates **boolean filtering** in Pandas using a column condition.

---

### **Finding the Largest LEGO Sets by Number of Parts**

To identify the LEGO sets with the most parts, we sort by the `num_parts` column in **descending order**.

```python
sets.sort_values('num_parts', ascending=False).head()
```

#### **Top LEGO Sets by Part Count**

1. **The Ultimate Battle for Chima** — 9,987 parts
2. **UCS Millennium Falcon** — 7,541 parts
3. **Hogwarts Castle** — 6,020 parts
4. **Taj Mahal (2017)** — 5,923 parts
5. **Taj Mahal (2008)** — 5,922 parts

---

### **Key Takeaways**

* `.sort_values()` defaults to ascending order
* Use `ascending=False` to find maximum values
* Boolean filtering (`df[column] == value`) is essential for targeted analysis
* LEGO’s product range has grown from **5 sets** to **massive multi-thousand-piece builds**
