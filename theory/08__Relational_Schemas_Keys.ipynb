## 8)— **Relational Database Schemas: Primary and Foreign Keys**

### **Motivation**

LEGO has licensed many major franchises (Star Wars, Harry Potter, Marvel, etc.) and also produces its own original themes.

A natural question arises:

**Which LEGO theme contains the largest number of individual sets?**

To answer this, we must understand how LEGO’s datasets are **linked together using relational database concepts**.

---

### **Counting the Number of Sets per Theme**

We start by counting how many sets belong to each theme using the `theme_id` column in `sets.csv`.

```python
set_theme_count = sets['theme_id'].value_counts()
set_theme_count[:5]
```

This tells us which `theme_id` appears most frequently.

#### **Problem**

The output only shows **numeric theme IDs**, not human-readable names.

Example:

* Theme ID `158` → 753 sets

This is not useful unless we know **what theme ID 158 represents**.

---

### **Why We Need Relational Data**

The LEGO dataset is **relational**:

* `sets.csv` contains:

  * `theme_id` (foreign key)

* `themes.csv` contains:

  * `id` (primary key)
  * `name` (theme name)

To map theme IDs to theme names, we must use **both tables**.

---

### **Understanding `themes.csv`**

The `themes.csv` file contains the **actual theme names**.

We load and inspect the dataset:

```python
themes = pd.read_csv('data/themes.csv')
themes.head()
```

#### **Key Columns**

* `id` → unique theme identifier (primary key)
* `name` → theme name
* `parent_id` → hierarchical relationship between themes

The `id` column is **unique**, making it the **primary key**.

---

### **Primary vs Foreign Keys (Concrete Example)**

* `themes.id` → **Primary Key**
* `sets.theme_id` → **Foreign Key**

This means:

* Many sets can belong to one theme
* Each theme is uniquely identified in `themes.csv`

---

### **Exploring a Real Example: Star Wars**

Searching for all themes named `"Star Wars"`:

```python
themes[themes.name == 'Star Wars']
```

#### **Result**

There are **4 different theme IDs** associated with the name *Star Wars*.

This explains why a franchise can have:

* Multiple sub-themes
* Different product lines over time

---

### **Linking Theme IDs Back to Sets**

Using one of the Star Wars theme IDs (e.g. `18`), we find the corresponding sets:

```python
sets[sets.theme_id == 18]
```

These sets include:

* Character-based sets (R2-D2, Darth Vader, Stormtrooper, etc.)
* Released primarily between 2000–2002

Checking another Star Wars theme ID:

```python
sets[sets.theme_id == 209]
```

These correspond to:

* Star Wars Advent Calendars
* All sharing the same `theme_id`
* Released annually over multiple years

---

### **Why Multiple Theme IDs Exist**

Large franchises like Star Wars:

* Span decades
* Contain multiple product categories
* Are subdivided into distinct themes internally

This design:

* Keeps data organised
* Allows LEGO to track product lines precisely

---

### **Key Takeaways**

* Relational databases split information across tables intentionally
* `themes.id` is the primary key
* `sets.theme_id` is the foreign key
* Theme names are **not unique**, but theme IDs are
* Understanding schema structure is essential before merging data
* Large franchises often span multiple theme IDs
