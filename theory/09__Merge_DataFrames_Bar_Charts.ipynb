## **09)—How to Merge DataFrames and Create Bar Charts**

### **Objective**

Combine LEGO **theme names** with the **number of sets per theme**, then visualise the top themes using a **bar chart**.

This section introduces:

* Converting a Pandas Series into a DataFrame
* Merging DataFrames using a **primary key**
* Creating and customising **bar charts** with Matplotlib

---

### **Context**

So far, we know:

* `sets.csv` contains LEGO sets and their `theme_id`
* `themes.csv` contains theme names with a unique `id`
* We previously computed how many sets belong to each `theme_id`

What we *don’t* have yet:

* A table that shows **theme name + number of sets together**

This is where **DataFrame merging** becomes essential.

---

### **Step 1: Count Number of Sets per Theme**

We start with a Pandas Series created using `value_counts()`:

```python
set_theme_count = sets["theme_id"].value_counts()
set_theme_count[:5]
```

This gives us:

* Index → `theme_id`
* Values → number of sets

However, this is still a **Series**, not a DataFrame.

---

### **Step 2: Convert Series into a DataFrame**

To prepare for merging, we create a proper DataFrame and name the columns explicitly:

```python
set_theme_count = pd.DataFrame({
    "id": set_theme_count.index,
    "set_count": set_theme_count.values
})

set_theme_count.head()
```

Now we have:

* `id` → theme identifier
* `set_count` → number of sets per theme

This matches the structure of `themes.csv`.

---

### **Step 3: Merge with themes.csv**

Both DataFrames share the column `id`, which acts as the **primary key** in `themes.csv`.

```python
merged_df = pd.merge(set_theme_count, themes, on="id")
merged_df.head(3)
```

Resulting columns:

* `id`
* `set_count`
* `name`
* `parent_id`

We now have theme names attached to their set counts.

---

### **Key Insight**

At this point, we can already see that **Star Wars** is the theme with the **highest number of LEGO sets**.

---

### **Step 4: Create a Basic Bar Chart**

To visualise the top 10 themes:

```python
plt.bar(merged_df.name[:10], merged_df.set_count[:10])
```

This works, but the chart is **hard to read** due to overlapping labels.

---

### **Step 5: Improve Chart Readability**

We apply standard Matplotlib customisation:

```python
plt.figure(figsize=(14, 8))
plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)
plt.xlabel("Theme Name", fontsize=14)
plt.ylabel("Nr of Sets", fontsize=14)

plt.bar(merged_df.name[:10], merged_df.set_count[:10])
```

Now the chart is:

* Wide enough
* Properly labelled
* Readable

---

### **Final Interpretation**

From the bar chart we observe:

* **Star Wars** has the most LEGO sets overall
* Other major themes include:

  * Gear
  * Friends
  * Ninjago
  * City / Town
* LEGO produces a large number of **non-traditional products** such as:

  * Books
  * Key chains
  * Accessories

This highlights LEGO’s **diversification beyond bricks**.

---

### **Key Takeaways**

* `.value_counts()` is useful but often needs conversion to a DataFrame
* `pd.merge()` allows relational-style joins in Pandas
* Clean visualisation requires:

  * Figure sizing
  * Axis labels
  * Rotated tick labels
* Data storytelling improves dramatically when raw numbers are combined with labels
