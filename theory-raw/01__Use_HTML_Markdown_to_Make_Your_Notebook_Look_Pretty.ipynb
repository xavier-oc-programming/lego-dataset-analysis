## **1)— Use HTML Markdown to Make Your Notebook Look Pretty**

### **Overview**

In this section, you learn how to improve the **visual structure and readability** of your Jupyter Notebook by combining **Markdown** with **HTML**. While the starter notebook already includes Markdown cells, using HTML allows you to further customise layout, headings, and images.

---

### **Notebook Cell Types**

A Jupyter Notebook contains two main types of cells:

* **Code cells**
  Used to write and execute Python code.

* **Text (Markdown) cells**
  Used for explanations, section titles, instructions, and challenges.

The starter notebook already includes several Markdown cells with headings and challenge descriptions, which you can enhance further using HTML.

---

### **Insert a Markdown Cell**

* Add a new **Text (Markdown)** cell below the *Introduction* section.
* This cell can be used for:

  * Section headings
  * Descriptions
  * Images
  * Structured explanations

Markdown cells support:

* Standard Markdown syntax
* Raw HTML tags

---

### **Adding Images**

You can display images inside a Markdown cell using the HTML `<img>` tag.

**Using an image from a URL:**

```html
<img src="https://i.imgur.com/49FNOHj.jpg">
```

**Using a local image file (VS Code / local Jupyter):**

```html
<img src="assets/bricks.jpg">
```

Notes:

* Local images work as long as the relative path is correct.
* Image assets are included in the lesson ZIP.
* When running locally, you can reference these files directly.

---

### **Section Headings**

You can create section headings in two ways:

**HTML headings:**

```html
<h1>Heading 1</h1>
<h2>Heading 2</h2>
```

**Markdown shorthand:**

```markdown
# Heading 1
## Heading 2
### Heading 3
#### Heading 4
##### Heading 5
```

Notes:

* Markdown headings scale down to a minimum of `#####`.
* Markdown shorthand is usually preferred for readability.
* HTML is useful when you need finer control over presentation.

---

### **Why This Matters**

Combining Markdown and HTML allows you to:

* Structure notebooks clearly
* Improve readability
* Make analyses easier to follow
* Produce notebooks that look polished and professional
