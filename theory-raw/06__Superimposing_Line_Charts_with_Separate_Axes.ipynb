## 6)— **Superimposing Line Charts with Separate Axes**

### **The Problem with Different Scales**

We want to compare:

* **Number of LEGO sets per year**
* **Number of LEGO themes per year**

If we plot both lines on the **same y-axis**, the result is misleading because:

* Themes range roughly from **0 to 90**
* Sets range roughly from **0 to 900**

Plotting them together without adjustment flattens the smaller-scale data and hides useful information.

```python
plt.plot(themes_by_year.index[:-2], themes_by_year.nr_themes[:-2])
plt.plot(sets_by_year.index[:-2], sets_by_year.set_num[:-2])
```

This chart is technically correct but visually uninformative.

---

### **Solution: Two Separate Y-Axes**

To properly compare both trends, we need **two y-axes** that share the same x-axis (year).

This is done using Matplotlib’s `.twinx()` method.

---

### **Creating Two Axes Objects**

```python
ax1 = plt.gca()      # get current axis
ax2 = ax1.twinx()   # create a second y-axis sharing the same x-axis
```

* `ax1` controls the first y-axis
* `ax2` controls the second y-axis
* Both axes share the same x-axis (years)

---

### **Plotting Data on Separate Axes**

```python
ax1.plot(sets_by_year.index[:-2], sets_by_year.set_num[:-2])
ax2.plot(themes_by_year.index[:-2], themes_by_year.nr_themes[:-2])
```

This allows each dataset to use its **own scale**, making both trends visible and meaningful.

---

### **Improving Readability with Styling**

Because both lines default to the same colour, we must style them clearly.

```python
ax1 = plt.gca()
ax2 = ax1.twinx()

# Plot with styling
ax1.plot(
    sets_by_year.index[:-2],
    sets_by_year.set_num[:-2],
    color='green'
)

ax2.plot(
    themes_by_year.index[:-2],
    themes_by_year.nr_themes[:-2],
    color='blue'
)

# Axis labels
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Sets', color='green')
ax2.set_ylabel('Number of Themes', color='blue')
```

---

### **What This Chart Shows**

* The **number of sets** grows much faster than the **number of themes**
* LEGO increases product output primarily by:

  * Expanding within existing themes
  * Releasing more sets per theme
* Both trends accelerate strongly after the mid-1990s

---

### **Key Takeaways**

* Never plot datasets with vastly different scales on a single axis
* `.twinx()` allows meaningful comparisons on the same chart
* Axis colouring should match the plotted data
* Separate axes preserve interpretability without distorting trends
