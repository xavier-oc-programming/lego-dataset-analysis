## 5)— **How to Use the Pandas `.agg()` Function**

### **Why `.agg()` Is Useful**

While `.groupby()` is excellent for splitting data into groups, we often want to apply **specific aggregation operations** to one or more columns.
This is where the Pandas `.agg()` method becomes essential.

In this case, we want to:

* Group LEGO sets by **year**
* Count how many **unique themes** (`theme_id`) existed in each year

---

### **Counting Themes per Calendar Year**

We chain `.groupby()` and `.agg()` together to calculate the number of unique themes released per year.

```python
themes_by_year = sets.groupby('year').agg({
    'theme_id': pd.Series.nunique
})
```

The `.agg()` method takes a **dictionary**:

* Key → column name
* Value → aggregation function to apply

Here, we use `.nunique()` to count distinct theme IDs.

---

### **Renaming the Aggregated Column**

To make the DataFrame clearer and more readable, we rename the column.

```python
themes_by_year.rename(
    columns={'theme_id': 'nr_themes'},
    inplace=True
)

themes_by_year.head()
```

### **Important Note About Incomplete Years**

As with the sets-per-year analysis:

* 2020 is incomplete
* 2021 contains only partial data

These years must be excluded from any time-series visualisation to avoid misleading trends.

---

### **Visualising the Number of Themes Over Time**

To plot only **full calendar years**, we use slicing to remove the last two rows.

```python
plt.plot(
    themes_by_year.index[:-2],
    themes_by_year.nr_themes[:-2]
)
```

* LEGO steadily increased the number of themes until the **mid-1990s**
* After that, theme growth **flattened** for roughly a decade
* A new expansion phase began in the early 2010s
* Theme variety grew alongside LEGO’s global expansion and product diversification

### **Key Takeaways**

* `.agg()` allows fine-grained control over grouped calculations
* Aggregation dictionaries make transformations explicit and readable
* `.nunique()` is ideal for counting distinct categories
* Always remove incomplete years when analysing time-series data
* The number of themes grew more slowly than the number of sets
