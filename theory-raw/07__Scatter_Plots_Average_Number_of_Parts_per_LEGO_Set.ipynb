## 7)— **Scatter Plots: Average Number of Parts per LEGO Set**

### **Complexity Over Time**

Have LEGO sets become **larger and more complex** over time?

To answer this, we calculate the **average number of parts per LEGO set per year**.
This is another practical use of combining `.groupby()` with `.agg()`.

---

### **Challenge: Create `parts_per_set`**

We want a Pandas object where:

* The **index** is the year
* The **values** are the average number of parts per LEGO set in that year

The result should look like this when inspected with `.head()` and `.tail()`.

---

### **Solution: Using `.groupby()` and `.agg()` Together**

We group the dataset by `year` and apply the `mean()` function to the `num_parts` column.

```python
parts_per_set = sets.groupby('year').agg({
    'num_parts': pd.Series.mean
})
```

This creates a DataFrame where:

* Each row represents a year
* `num_parts` represents the **average number of parts per set**

---

### **Inspecting the Results**

```python
parts_per_set.head()
```

#### **Early Years**

* 1949 → ~99.6 parts
* 1950 → 1.0 part
* 1953 → ~13.5 parts
* 1954 → ~12.4 parts
* 1955 → ~36.6 parts

```python
parts_per_set.tail()
```

#### **Recent Years**

* 2017 → ~221.8 parts
* 2018 → ~213.6 parts
* 2019 → ~207.5 parts
* 2020 → ~259.7 parts
* 2021 → 0.0 parts (incomplete data)

---

### **Important Data Quality Note**

As in previous sections:

* 2020 is an incomplete year
* 2021 contains invalid averages

These rows must be excluded before plotting.

---

### **Visualising with a Scatter Plot**

A scatter plot is ideal because:

* Each point represents a single year
* Long-term trends remain visible without smoothing

```python
plt.scatter(
    parts_per_set.index[:-2],
    parts_per_set.num_parts[:-2]
)
```

---

### **Interpreting the Scatter Plot**

From the chart, we can clearly observe:

* A strong **upward trend** in average LEGO set size
* In the 2010s, the average LEGO set contained around **200 parts**
* This is roughly **double** the average size of LEGO sets in the 1960s

This confirms that LEGO sets have become **larger and more complex** over time.

---

### **Key Takeaways**

* `.agg()` allows flexible aggregation logic
* `.mean()` is useful for analysing complexity trends
* Scatter plots are effective for time-based comparisons
* Always exclude incomplete years from time-series analysis
* LEGO sets have steadily increased in size over decades
